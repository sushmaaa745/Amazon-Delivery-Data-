```python
missing_values = data.isnull().sum()
print(f"Missing values:\n{missing_values}")
```

```python
data = data.drop(columns=data.columns[data.columns.duplicated()])
print("Columns:\n", data.columns)
```

```python
data['source_name'].fillna('Unknown', inplace=True)
data['destination_name'].fillna('Unknown', inplace=True)
missing_values_after = data.isnull().sum()
print(f"Missing values after filling:\n{missing_values_after}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv( r'C:\Users\HP\OneDrive\Desktop\delhivery_data.csv')
sns.set(style="whitegrid")
fig, axes = plt.subplots(3, 2, figsize=(15, 15))
sns.histplot(data['actual_time'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Actual Time')
sns.histplot(data['osrm_time'], kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of OSRM Time')
sns.histplot(data['factor'], kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Factor')
sns.histplot(data['segment_actual_time'], kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Segment Actual Time')
sns.histplot(data['segment_osrm_time'], kde=True, ax=axes[2, 0])
axes[2, 0].set_title('Distribution of Segment OSRM Time')
sns.histplot(data['segment_factor'], kde=True, ax=axes[2, 1])
axes[2, 1].set_title('Distribution of Segment Factor')
plt.tight_layout()
plt.show()
```

```python
#categorical variables
plt.figure(figsize=(10, 6))
sns.countplot(y='source_name', data=data, order=data['source_name'].value_counts().index)
plt.title('Count of Source Names')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(y='destination_name', data=data, order=data['destination_name'].value_counts().index)
plt.title('Count of Destination Names')
plt.show()
```

```python
# Scatterplots for numerical variables
plt.figure(figsize=(8, 6))
sns.scatterplot(x='actual_time', y='osrm_time', data=data)
plt.title('Scatter Plot between Actual Time and OSRM Time')
plt.show()

#categorical vs numerical variables
plt.figure(figsize=(12, 8))
sns.boxplot(x='source_name', y='actual_time', data=data)
plt.title('Box Plot of Actual Time by Source Name')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='destination_name', y='actual_time', data=data)
plt.title('Box Plot of Actual Time by Destination Name')
plt.xticks(rotation=90)
plt.show()
```

```python
# Convert trip_creation_time to datetime
data['trip_creation_time'] = pd.to_datetime(data['trip_creation_time'])

# Extracting year, month, day from trip_creation_time
data['year'] = data['trip_creation_time'].dt.year
data['month'] = data['trip_creation_time'].dt.month
data['day'] = data['trip_creation_time'].dt.day

#trips per year
plt.figure(figsize=(10, 6))
sns.countplot(x='year', data=data)
plt.title('Number of Trips per Year')
plt.show()

#trips per month
plt.figure(figsize=(10, 6))
sns.countplot(x='month', data=data)
plt.title('Number of Trips per Month')
plt.show()

#trips per day
plt.figure(figsize=(10, 6))
sns.countplot(x='day', data=data)
plt.title('Number of Trips per Day')
plt.show()
```

```python
#to identify numerical columns 
plt.figure(figsize=(12, 8))
sns.boxplot(data=data[['actual_time', 'osrm_time', 'factor', 'segment_actual_time', 'segment_osrm_time', 'segment_factor']])
plt.title('Box Plot to Identify Outliers')
plt.xticks(rotation=90)
plt.show()
```

```python
#analyzing relationships
sns.pairplot(data[['actual_time', 'osrm_time', 'factor', 'segment_actual_time', 'segment_osrm_time', 'segment_factor']])
plt.show()
```
